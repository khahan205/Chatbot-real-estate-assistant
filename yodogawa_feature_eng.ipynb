```python
import pandas as pd
import numpy as np
```

```python
def load_data(file_path):
    df = pd.read_csv(file_path)
    return df
```

```python
def feature_engineering(df):

    # PRICE PER M2
    df["price_per_m2"] = df["rent"] / df["area_m2"]

    # TOTAL COST
    if "management_fee" in df.columns:
        df["total_cost"] = df["rent"].fillna(0) + df["management_fee"].fillna(0)
    else:
        df["total_cost"] = df["rent"].fillna(0)

    # AGE GROUP
    def age_group(age):
        if age < 5:
            return "new"
        elif age < 15:
            return "medium"
        else:
            return "old"

    df["age_group"] = df["building_age"].apply(age_group)

    # DISTANCE SCORE
    df["distance_score"] = 1 / (df["walking_distance_to_station"] + 1)

    # SIZE CATEGORY
    def size_category(area):
        if area < 25:
            return "small"
        elif area < 50:
            return "medium"
        else:
            return "large"

    df["size_category"] = df["area_m2"].apply(size_category)

    # COST PERFORMANCE
    df["cost_performance"] = df["area_m2"] / df["total_cost"].replace(0, np.nan)

    return df


```

```python
def feature_engineering(df):

    # 🔥 FIX TÊN CỘT
    df.rename(columns={
        "賃料": "rent",
        "家賃": "rent",
        "price": "rent",

        "専有面積": "area_m2",
        "area": "area_m2",

        "築年数": "building_age",
        "age": "building_age",

        "駅徒歩": "walking_distance_to_station",
        "distance": "walking_distance_to_station"
    }, inplace=True)
    return df
```

```python
required_cols = ["rent", "area_m2", "building_age", "walking_distance_to_station"]

for col in required_cols:
    if col not in df.columns:
        print(f"❌ Missing column: {col}")
```

```python
if __name__ == "__main__":

    input_path = r".\data\yodogawa_data_cleaning.csv"
    output_path = r".\data\yodogawa_feature_eng.csv"

    df = load_data(input_path)

    df = feature_engineering(df)

    print("Feature engineering completed!")
    print(df.head())

    df.to_csv(output_path, index=False)
```
